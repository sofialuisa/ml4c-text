### Usando Markov chains para generar texto

Utilizaremos la libreria creada por Allison Parrish shmarkov.py: [disponible aquí](https://gist.github.com/aparrish/14cb94ce539a868e6b8714dd84003f06) 
(referenciada como `shmarkov.py`)

```python
from shmarkov import *
```

Creating 3 types of character level markov chains with different temperatures. 

```python
#-------------------------------------------------------------------------*-------tutexto.txt.-------------*--char/word
contenido1 = [line.strip() for line in markov_generate_from_lines_in_file(2, open("radiohead_lyrics.txt"), 100, "char")]
contenido2 = [line.strip() for line in markov_generate_from_lines_in_file(4, open("radiohead_lyrics.txt"), 100, "word")]
contenido3 = [line.strip() for line in markov_generate_from_lines_in_file(5, open("radiohead_lyrics.txt"), 100, "char")]
```

Combining all of the different temperatures results into one.

```python
print(contenido1)
```

```python
resultados = contenido1 + contenido2 + contenido3
```

Getting random results to make a selection of results later. I'll choose 30 interesting phrases out of 300. 

```python
import random
import textwrap
```

```python
print('\n'.join(random.sample(resultados, 10)))
```
